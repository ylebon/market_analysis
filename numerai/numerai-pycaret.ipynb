```python
# import dependencies
import pandas as pd
import numerapi
import sklearn.linear_model
from pycaret.regression import *
```

```python
# download the latest training dataset (takes around 30s)
training_data = pd.read_csv("https://numerai-public-datasets.s3-us-west-2.amazonaws.com/latest_numerai_training_data.csv.xz")
training_data.head()
```

```python
# find only the feature columns
feature_cols = training_data.columns[training_data.columns.str.startswith('feature')]
feature_cols = feature_cols.tolist()
```

```python
# select those columns out of the training dataset
cols = feature_cols + ['target']
train_data = training_data[cols]
```

```python
reg101 = setup(data = train_data, target = 'target', session_id=123)
```

```python
del training_data
```

```python
best = compare_models(exclude = ['br', 'lasso', 'en', 'llar', 'knn', 'dt', 'par', 'huber', 'rf', 'et', 'gbr'])
```

```python
model_name = "lightgbm"
```

```python
model = create_model(model_name)
```

```python
#tuned_model = tune_model(model)
tuned_model = model
```

```python
plot_model(tuned_model)
```

```python
plot_model(tuned_model, plot='error')
```

```python
plot_model(tuned_model, plot='feature')
```

```python
predict_model(tuned_model);
```

```python
final_model = finalize_model(tuned_model)
```

```python
tournament_data = pd.read_csv("https://numerai-public-datasets.s3-us-west-2.amazonaws.com/latest_numerai_tournament_data.csv.xz")
```

```python
predictions=predict_model(final_model, data=tournament_data[feature_cols])
```

```python
# predictions must have an `id` column and a `prediction` column
predictions_df = tournament_data["id"].to_frame()
predictions_df["prediction"] = predictions["Label"]
predictions_df.head()
```
